```python
import numpy as np

%config InlineBackend.figure_formats = ['svg']
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
```

```python
def circle(c,r,N):
    θ=np.random.uniform(low=0.0, high=2*np.math.pi, size=N)
    θ=np.linspace(0,2*np.math.pi,N)
    x=r*np.cos(θ)+c[0]
    y=r*np.sin(θ)+c[1]
    return x,y
def prog(x,y,c,ϕ):
    return np.cos(ϕ)*(y-c[1])+np.sin(ϕ)*(x-c[0])
```

```python
C=[0,0]
N=30
S=30
ϕ=np.linspace(0,2*np.pi,S)
x,y=circle(C,1,N)
j,k=circle(C,2,N)
z=np.zeros((S,N))
z1=np.zeros((S,N))
plt.figure(figsize=(6, 1.8))
plt.subplot(1, 2, 1)
for i in range(0,S):
    z[i,:]=prog(x,y,C,ϕ[i])
    plt.scatter(np.repeat(ϕ[i],N),z[i,:], marker='s',color="black",s=1.5,alpha=0.3)
    z1[i,:]=prog(j,k,C,ϕ[i])
    plt.scatter(np.repeat(ϕ[i],N),z1[i,:],marker='^',color="red",s=1.5,alpha=0.3)
plt.tick_params(axis='both', labelsize=13)
plt.xlabel(r"$\phi$",fontsize=13)
plt.ylabel(r"$z_{1}$",fontsize=13)
plt.yticks([-2,-1,0,1,2])
plt.xticks([0,1/2*np.pi,1*np.pi,3/2*np.pi,2*np.pi],[r"$0$","$\pi/2$","$\pi$","$3/2\pi$","$2\pi$"])
plt.tight_layout()
plt.annotate(r"$a)$", (-2.0, 1.7), fontsize="x-large", annotation_clip=False)

plt.subplot(1, 2, 2)
β=np.random.rand()*2*np.pi
plt.xlabel(r"$\theta$",fontsize=13)
plt.ylabel(r"$z_{2}$",fontsize=13)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks([0,1/2*np.pi,1*np.pi,3/2*np.pi,2*np.pi],[r"$0$","$\pi/2$","$\pi$","$3/2\pi$","$2\pi$"])
for a in ϕ:
    h1=prog(x,y,C,β)*np.cos(a)+np.sqrt((x-C[0])**2+(y-C[1])**2)*np.sin(a)
    plt.scatter(np.repeat(a,N),h1,marker='s',color="black",s=1.5,alpha=0.3)
    h2=prog(j,k,C,β)*np.cos(a)+np.sqrt((j-C[0])**2+(k-C[1])**2)*np.sin(a)
    plt.scatter(np.repeat(a,N),h2,marker='^',color="red",s=1.5,alpha=0.3)
    if max(h1)<min(h2) or max(h2)<min(h1):
        plt.axvline(x=a,linewidth=0.8, color='gray',linestyle=':')
        print(a,β)
plt.annotate(r"$b)$", (-2.0, 2.5), fontsize="x-large", annotation_clip=False)
plt.tight_layout()
plt.savefig("toy_example.pdf")
```
